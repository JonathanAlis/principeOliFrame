```python
from moviepy import VideoFileClip, clips_array, ColorClip, CompositeVideoClip
import numpy as np
# Carregar seus 8 vídeos
_videos = [
    VideoFileClip("results/fire_n_flames_part_0.mp4"),
    VideoFileClip("results/fire_n_flames_part_2.mp4"),
    VideoFileClip("results/fire_n_flames_part_3.mp4"),
    VideoFileClip("results/fire_n_flames_part_4.mp4"),
    VideoFileClip("results/fire_n_flames_part_6.mp4"),
]

target_height = 1920 // 5 *2    # cada video ocupa metade da altura da coluna
target_width = 1080 // 2     # cada video ocupa 1/4 da largura do restante pra 1080

_videos = [v.resized(width=target_width) for v in _videos]
_videos[4].cropped(width=target_width, height=target_height//2, 
              x_center=target_width, y_center=target_height)
# Redimensiona mantendo proporção
def resize_group(videos):
    if target_width/target_height > videos[0].w/videos[0].h:
        return [v.resized(width=target_width) for v in videos]
    else:
        return [v.resized(height=target_height) for v in videos]
def black_mask(get_frame, t):
    frame = get_frame(t)
    # se for quase preto, máscara = 0 (transparente)
    if np.mean(frame) < 5:
        return np.zeros((frame.shape[0], frame.shape[1]), dtype=np.uint8)
    else:
        return np.ones((frame.shape[0], frame.shape[1]), dtype=np.uint8) * 255


clips = clips_array([[_videos[0], _videos[1]],
                        [_videos[2], _videos[3]]])
overlay = _videos[4].with_position(("center", "center"))

mask_clip = overlay.transform(lambda gf, t: black_mask(gf, t))
overlay = overlay.with_mask(mask_clip)

final = CompositeVideoClip([clips, overlay])

final.write_videofile("results/fire_n_flames.mp4", fps=30)
```

```python
from moviepy import VideoFileClip, clips_array, CompositeVideoClip

# parâmetros do frame final
final_w, final_h = 1080, 1920

# vídeos
top = [VideoFileClip("results/fire_n_flames_part_0.mp4"), 
       VideoFileClip("results/fire_n_flames_part_2.mp4")]
middle = VideoFileClip("results/fire_n_flames_part_6.mp4")
bottom = [VideoFileClip("results/fire_n_flames_part_3.mp4"), 
       VideoFileClip("results/fire_n_flames_part_4.mp4")]
h_top = h_bottom = int(final_h * 2/5)
h_middle = int(final_h * 1/5)
w_top = w_bottom = final_w // 2

def resize_crop(clip, target_w, target_h):
    if clip.w / clip.h > target_w / target_h:
        clip = clip.resized(height=target_h)
    else:
        clip = clip.resized(width=target_w)
    clip = clip.cropped(width=target_w, height=target_h,
                        x_center=clip.w//2, y_center=clip.h//2)
    return clip

top = [resize_crop(v, w_top, h_top) for v in top]
bottom = [resize_crop(v, w_bottom, h_bottom) for v in bottom]
middle = resize_crop(middle, final_w, h_middle)

# define posições manualmente
top[0] = top[0].with_position((0,0))
top[1] = top[1].with_position((w_top,0))
middle = middle.with_position((0,h_top))
bottom[0] = bottom[0].with_position((0,h_top+h_middle))
bottom[1] = bottom[1].with_position((w_bottom,h_top+h_middle))

all_clips = top + [middle] + bottom

final = CompositeVideoClip(all_clips, size=(final_w, final_h))

final.write_videofile("fire_and_flames.mp4", fps=30)
```
